```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read paths.csv
pathDF = pd.read_csv('data/dataset/paths.csv', encoding='utf-8')
pathDF['length'] = pathDF.urls.apply(lambda xx: len(xx.split('-')))
pathDF
```

```python
# Count of sessions
pathDF.counts.sum()
```

```python
# data clean

minThreshold, maxThreshold = 10, 300

groupbyDF = pathDF.groupby(by=['length'])
groupDict = groupbyDF.groups.copy()
lengthOfSession, countOfLS = [], []

for length in groupDict.keys():
    indexs = groupDict[length]
    t = pathDF.counts[indexs].sum() 
    if t < minThreshold or t > maxThreshold:
        pathDF = pathDF.drop(indexs)
    else:
        lengthOfSession.append(length)
        countOfLS.append(t)
pathDF
```

```python
# %matplotlib notebook
# fig, axes = plt.subplots(2, 2, sharex=False, sharey=False)

# axes[0, 0].scatter(dic.keys(), dic.values())

# plt.subplots_adjust(wspace=1, hspace=1)

fig = plt.figure()

ax = fig.add_subplot(1, 1, 1)
fig.set_label("the count of length of session")

ax.set_xlabel('length of session')
ax.set_ylabel('count')
ax.scatter(lengthOfSession, countOfLS)
```

```python
# pathDF.counts[[0,1,2,3,4]].sum()
rawData = []
splitData = []
for tup in pathDF.itertuples():
    rawData.extend([tup.urls for i in range(tup.counts)])
    splitData.extend([
        [int(j) for j in tup.urls.split('-')] for i in range(tup.counts)
    ])
```

```python
# regenerate dataset
splitData
```

```python
inputData = np.array(splitData)
np.save('data/dataset/train.npy', inputData)
```
